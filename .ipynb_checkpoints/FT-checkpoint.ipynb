```python
#フーリエ変換後のG_iwn

import numpy as np
import matplotlib.pyplot as plt

#逆温度
beta=100

#ω_n
n=np.arange(-100,101)
w_n=np.pi*(2*n+1)/beta

#G_iwn
G_iwn=1/(1J*w_n-1)

#グラフ
plt.plot(w_n, G_iwn.real,label='real', marker=".")
plt.plot(w_n, G_iwn.imag,label='imag', marker=".")
plt.xlabel('ωn')
plt.ylabel('G(iωn)')
plt.legend()
plt.show()
```

```python
#G_tをフーリエ変換してG_iwnの配列を作る

import numpy as np
import matplotlib.pyplot as plt

#逆温度
beta=100

#ω_n
n=np.arange(-100,101)
w_n=np.pi*(2*n+1)/beta
#0からbetaをN分割
N=1001
t=np.linspace(0,beta,N)#0からbetaまでの１００1個の配列


#G_iwn
g_matrix=np.fromfunction(lambda i,j:np.exp(1J*np.pi*(2*(i-100)+1)*(j+1/2)/(N-1)) ,(201,N-1))
h_vector=-np.exp(-beta*(np.arange(N-1)+1/2)/(N-1))/(1+np.exp(-beta))+1/2
G_iwn=beta/(N-1)*np.dot(g_matrix,h_vector)+1/(1j*w_n)

#グラフ
plt.plot(w_n, G_iwn.real,label='real', marker=".")
plt.plot(w_n, G_iwn.imag,label='imag', marker=".")
plt.xlabel('ωn')
plt.ylabel('G(iωn)')
plt.legend()
plt.show()
```

```python
#G_tをフーリエ変換してG_iwnの配列を作る。cubicsplineを使用。
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

#逆温度
beta=100

#ω_n
n=np.arange(-100,101)
W_n=np.pi*(2*n+1)/beta

#0からbetaをN分割
N=1001
t=np.linspace(0,beta,N)#0からbetaまでの１００1個の配列

#G_iwn
g_matrix=np.fromfunction(lambda i,j:np.exp(1J*np.pi*(2*(i-100)+1)*(j+1/2)/(N-1)) ,(201,N-1))
h_vector=-np.exp(-beta*(np.arange(N-1)+1/2)/(N-1))/(1+np.exp(-beta))+1/2
G_iwn=beta/(N-1)*np.dot(g_matrix,h_vector)+1/(1j*w_n)

w_n=np.arange(-6.25176938,6.31460123,0.05)
cubic = interp1d(W_n, G_iwn, kind="cubic")
plt.plot(w_n, cubic(w_n).real,label='real', marker=".")
plt.plot(w_n, cubic(w_n).imag,label='imag', marker=".")
plt.xlabel('ωn')
plt.ylabel('G(iωn)')
plt.legend()
plt.show()

#グラフ
```

```python
#フーリエ変換後のG_iwnとG_tをフーリエ変換したG_iwnを比較。

import numpy as np
import matplotlib.pyplot as plt

#逆温度
beta=100

#ω_n
n=np.arange(-100,101)
w_n=np.pi*(2*n+1)/beta

#G_iwn
G_iwn=1/(1J*w_n-1)


#G_tをフーリエ変換したG_iwn
#0からbetaをN分割
N=10001
t=np.linspace(0,beta,N)#0からbetaまでの１００００1個の配列

#G_iwn
g_matrix=np.fromfunction(lambda i,j:np.exp(1J*np.pi*(2*(i-100)+1)*j/(N-1)) ,(201,N-1))
h_vector=-np.exp(-beta*np.arange(N-1)/(N-1))/(1+np.exp(-beta))+1/2
G_iwn_ft=beta/(N-1)*np.dot(g_matrix,h_vector)+1/(1j*w_n)

#グラフ
plt.plot(w_n, G_iwn.real,label='real', marker=".")
plt.plot(w_n, G_iwn.imag,label='imag', marker=".")
plt.plot(w_n, G_iwn_ft.real,label='real_ft', marker=".")
plt.plot(w_n, G_iwn_ft.imag,label='imag_ft', marker=".")
plt.xlabel('ωn')
plt.ylabel('G(iωn)')
plt.legend()
plt.show()
```

```python
#誤差の計算
G_iwn_difference=G_iwn-G_iwn_ft
plt.plot(w_n, G_iwn_difference.real,label='real_ft', marker=".")
plt.plot(w_n, G_iwn_difference.imag,label='imag_ft', marker=".")
plt.xlabel('ωn')
plt.ylabel('G_iwn_difference')
plt.legend()
plt.show()
print("誤差の最大値(実部）:{0}、（虚部):{1}".format(np.max(G_iwn_difference.real),np.max(G_iwn_difference.imag)))
```
